```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/midhu/OneDrive/Desktop/midhun/spotify.csv")
print("Original Data:\n", data)
nan_value = data.select_dtypes(include=[np.number])
mean = nan_value.mean()                            
nan_value_filled = nan_value.fillna(mean)         
data.update(nan_value_filled)               
print("\nCleaned Data:\n", data)
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=data,
    x='valence',
    y='year',
    hue='acousticness',
    marker='o'
)
plt.title('Valence VS Year', fontsize=16)
plt.xlabel('Valence', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.grid(True)
plt.show()
```
